```python
import matplotlib.pyplot as plt

# Initial Equity
init_equity = 2000

# Monthly added value
monthly_added_value = 100

# Months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Monthly developments
monthly_dev = [0, -1.91, 6.94, -1.99, 31.53, 4.85, 33.40, 3.51, -0.65, 17.66, 18.61, 8.52]

# Result
def calculate_developments(monthly_dev, init_equity, monthly_added_value, profit_sharing = True):
    monthly_res = []
    monthly_res_with_monthly_payments = []
    equity = init_equity
    for interest in monthly_dev:
        delta = equity * interest/100
        equity = equity + delta
        if delta > 0 and profit_sharing: # profit_sharing
            equity = equity - (delta * (25/100))
        #print(f"Delta in one month {equity}.")
        monthly_res.append(equity)
        equity = equity + monthly_added_value
        
    return monthly_res

monthly_res_with_monthly_payments = calculate_developments(monthly_dev, init_equity, 100)
monthly_res = calculate_developments(monthly_dev, init_equity, 0)

print("Net profit=", monthly_res[-1] - init_equity)
print("Net profit with montly payments=", monthly_res_with_monthly_payments[-1] - init_equity+(monthly_added_value*11))
# print(init_equity + monthly_added_value * 11)
plt.plot(months, monthly_res, marker="o")   # or use plt.bar(months, monthly_dev)
plt.plot(months, monthly_res_with_monthly_payments, marker="o", label="Test")
plt.xlabel("Month")
plt.ylabel("Equity")
plt.title("Monthly Equity")
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```
